```python
df=pd.read_csv('/content/consumers.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Smoker'].fillna('No',inplace=True)
df['Transportation_Method'].fillna('Walking',inplace=True)
df['Marital_Status'].fillna('Married',inplace=True)
df['Children'].fillna(0,inplace=True)
df['Occupation'].fillna('Student',inplace=True)
df['Budget'].fillna('Medium',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.to_csv('consumers_cleaned.csv',index=False)
```
